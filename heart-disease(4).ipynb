20.plot continues variables

```python
cate_val=[]
cont_val=[]
for column in data.columns:
    if data[column].nunique() <=10:
        cate_val.append(column)
    else:
        cont_val.append(column)
data.hist(cont_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```
